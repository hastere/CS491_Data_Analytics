```python
classLabels = dict()

classLabels[0] = "Guardrail_Barriers"
classLabels[1] = "RumbleStrips"
```

```python
import os
X_train = []
Y_train = []
```

```python
#TODO: Read images of class type 'RoadsideBarriers'
Path_of_the_image_folder = "D:\\Arpan\\DataScience\\Lab3\\Task2\\Train\\RumbleStrips"
#TODO Ends
for file in os.listdir(Path_of_the_image_folder):
    img = image.load_img(Path_of_the_image_folder+'\\'+file, target_size=(224, 224))
    x = image.img_to_array(img)
    #x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    X_train.append(x)
    Y_train.append(1)
```

```python
X_train =np.asarray(X_train)
```

```python
X_train.shape
```

```python
num_classes = 2
Y_train = np_utils.to_categorical(Y_train, num_classes)
from sklearn.utils import shuffle
X_train, Y_train = shuffle(X_train, Y_train, random_state=1000)
```

```python
image_input = Input(shape=(224, 224, 3))
model = ResNet50(input_tensor=image_input, include_top=True,weights='imagenet')
#model = ResNet50(include_top=True,weights=None,classes=3)
```

```python
last_layer = model.get_layer('avg_pool').output
x= Flatten(name='flatten')(last_layer)
out = Dense(num_classes, activation='softmax', name='output_layer')(x)
custom_resnet_model = Model(inputs=image_input,outputs= out)
custom_resnet_model.summary()
```

```python
#Get Test Images
#TODO: Read images of class type 'Guardrail_Barriers'
#Note: NoRoadSideProtection has class label 0
#Path_of_the_image_folder = "Provide_path_to_the_testfolder_NoRoadSideProtection_here"
Path_of_the_image_folder = "D:\\Arpan\\DataScience\\Lab3\\Task2\\Test\\Guardrail_Barriers"
#TODO Ends
fileNames = []
X_test_Guardrail_Barriers = []
for file in os.listdir(Path_of_the_image_folder):
    img = image.load_img(Path_of_the_image_folder+'\\'+file, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    X_test_Guardrail_Barriers.append(x)
    fileNames.append(file)

X_test_Guardrail_Barriers= np.asarray(X_test_Guardrail_Barriers)
```
